```python
#dictionary by linked list
class Node():
    def __init__(self,key,value):
        self.data=value
        self.key=key
        self.next=None
        self.prev=None


class L_Dictionary():
    def __init__(self):
        self.head=None
        self.tail=None
        self.n=0
        
    def insert_from_head(self,key,value):
        self.n+=1
        if self.head==None:
            NN=Node(key,value)
            self.head=NN
            self.tail=self.head
        else:
            NN=Node(key,value)
            NN.next=self.head
            self.head.prev=NN
            self.head=NN
    def __setitem__(self,key,value):
        return self.insert_from_head(key,value)
    
    
    def insert_from_tail(self,key,value):
        self.n+=1
        if self.head==None:
            NN=Node(key,value)
            self.head=NN
            self.tail=self.head
        else:
            NN=Node(key,value)
            self.tail.next=NN
            NN.prev=self.tail
            self.tail=NN
    def insert_from_middle(self,key,value,index):
        
        if index==self.n-1:
            self.insert_from_tail(key,value)
        elif index<self.n:
            self.n+=1
            temp=self.head
            for i in range(index):
                temp=temp.next

            NN=Node(key,value)
            NN.next=temp.next
            temp.next.prev=NN
            temp.next=NN
            NN.prev=temp
        else:
            return"index not found"
    
    
    def traverse_from_head(self):
        if self.head==None:
            return "empty list"
        else:
            temp=self.head
            while temp!=None:
                print(temp.key,":",temp.data," ",end=" ")
                temp=temp.next
    def traverse_from_tail(self):
        if self.head==None:
            return "empty list"
        else:
            temp=self.tail
            while temp!=None:
                print(temp.key,":",temp.data," ",end=" ")
                temp=temp.prev
                
    def size(self):
        return self.n
    
    def delete_head(self):
        if self.head==None:
            return "empty list"
        elif self.head.next==None:
            self.n-=1
            print(self.head.data)
            self.head=None
            self.tail=None
        else:
            self.n-=1
            print(self.head.data)
            self.head=self.head.next
            self.head.prev=None
            
    def delete_tail(self):
        if self.head==None:
            return "empty list"
        if self.n==1:
            self.delete_head()
        else:
            self.tail=self.tail.prev
            self.tail.next=None
            self.n-=1
            
            
    def delete_by_value(self,key):
        if self.head.key==key:
            self.delete_head()
            return
        elif self.tail.key==key:
            self.delete_tail()
            return
        if self.head==None:
            return"empty list"
        temp=self.head
        while temp.next!=None:
            if temp.next.key==key:
                break
            temp=temp.next
        if temp.next==None:
            return"value not found"
        else:
            temp.next=temp.next.next
            temp.next.prev=temp
            self.n-=1

    def delete_by_value_all(self,key):
        if self.head==None:
            return "empty list"
        else:
            temp=self.head
            c=0
            while temp!=None:
                if temp.key==key:
                    c+=1
                temp=temp.next
                
            for i in range(c):
                self.delete_by_value(key)
                
    def search(self,key):
        if self.head==None:
            return"empty list"
        else:
            c=0
            temp=self.head
            while temp!=None:
                if temp.key==key:
                    break
                temp=temp.next
                c+=1
            if temp==None:
                return "not found"
            else:
                return temp.data
    def __getitem__(self,key):
        return self.search(key)
```

```python
def initialize_parameters(layer_dims):
    np.random.seed(3)
    parameters = L_Dictionary()
    L = len(layer_dims)
    for l in range(1, L):
        parameters['W' + str(l)] = np.random.randn(layer_dims[l-1], layer_dims[l])*1/(layer_dims[l-1]**1/2) #xavier/glorat initializer
        parameters['b' + str(l)] = np.zeros((layer_dims[l], 1))
    return parameters
```

```python
# Utility Functions
def sigmoid(Z):
    A = 1/(1+np.exp(-Z))
    return A
def linear_forward(A_prev, W, b):

    Z = np.dot(W.T, A_prev) + b
    A=sigmoid(Z)
    return A
```

```python
# Forward Prop
def L_layer_forward(X, parameters):

    A = X
    L = parameters.n // 2                  # number of layers in the neural network

    for l in range(1, L+1):
        A_prev = A
        Wl = parameters['W' + str(l)]
        bl = parameters['b' + str(l)]
        # print("A"+str(l-1)+": ", A_prev)
        # print("W"+str(l)+": ", Wl)
        # print("b"+str(l)+": ", bl)
        # print("--"*20)

        A = linear_forward(A_prev, Wl, bl)
        # print("A"+str(l)+": ", A)
        # print("**"*20)

    return A,A_prev
```

```python
class L_Node():
    def __init__(self,value):
        self.data=value
        self.next=None
        
class Queue():
    def __init__(self):
        self.head=None
        self.tail=None
        self.n=0
        
    def enqueue(self,value):
        self.n+=1
        if self.head==None:
            NN=L_Node(value)
            self.head=NN
            self.tail=self.head
            
        else:
            NN=L_Node(value)
            self.tail.next=NN
            self.tail=NN
            
    def search_by_index(self,index):
        if self.head==None:
            return"empty list"
        else:
            temp=self.head
            for i in range(index):
                temp=temp.next
            # if temp==None:
            #     return "not found"
            return temp.data
    def __getitem__(self,index):
        return self.search_by_index(index)
        
    def traverse(self):
        if self.head==None:
            print("empty queue")
        else:
            temp=self.head
            while temp!=None:
                print(temp.data)
                temp=temp.next
                
    def size(self):
        return self.n
    
    def peek(self):
        if self.head==None:
            print("empty queue")
        else:
            print(self.head.data)
            
    def clear(self):
        self.n=0
        while self.head!=None:
            self.head=self.head.next
        
        
```
